```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree

data = {
    'Departamento': ['Ventas', 'Soporte', 'Logstica', 'Ventas', 'Soporte', 'Logstica'],
    'Antigedad': [2, 5, 3, 1, 4, 6],
    'Cumplimiento de Objetivos': [80, 60, 75, 50, 85, 70],
    'Recibe Bono': ['No', 'No', 'S', 'No', 'Si', 'No']
}

data['Recibe Bono'] = [1 if bono == 'S' else 0 for bono in data['Recibe Bono']]

df = pd.DataFrame(data)

X = df[['Antigedad', 'Cumplimiento de Objetivos']]
y = df['Recibe Bono']

cart_model = DecisionTreeClassifier(random_state=42)
cart_model.fit(X, y)

plt.figure(figsize=(12, 8))
plot_tree(cart_model, feature_names=X.columns, class_names=['No', 'S'], filled=True, rounded=True)
plt.show()
```
